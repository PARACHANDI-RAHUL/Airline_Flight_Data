**Show Bar Graphs representing the Departure Time & Arrival Time.**

```python
df['departure_time'].value_counts()
```

```python
df['arrival_time'].value_counts()
```

```python
plt.figure(figsize=(16,4))

plt.subplot(1,2,1)
df['departure_time'].value_counts(ascending=False).plot.bar(color=['y','g'])
plt.title('Departure Time Frequencies')
plt.xlabel('Departure Time')
plt.ylabel('Total departure flights')

plt.subplot(1,2,2)
df['arrival_time'].value_counts(ascending=False).plot.bar(color=['y','g'])
plt.title('Arrival_time Frequencies')
plt.xlabel('arrival Time')
plt.ylabel('Total arrival flights')
plt.show()
```
